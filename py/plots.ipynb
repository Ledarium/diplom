```python
import matplotlib as mpl
```

```python
s_const, s_var = 68269, 6000
c_const, c_var = 0, 17271
cli_max = 25
serv = [(s_const + s_var*i)/1000 for i in range(cli_max+1)]
pc = [(c_const + c_var*i)/1000  for i in range(cli_max+1)]

plt.xlim(0,cli_max)
plt.ylim(0, 500)
plt.grid(True)

plt.plot(serv, '-o')
plt.plot(pc, '-o')

plt.xlabel('Клиенты', fontsize=10)
plt.ylabel('Стоимость, тыс. руб.', fontsize=10)
plt.savefig('../img/hw_price_chart.png')
```
