```python
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
import matplotlib as mpl
mpl.rcParams['figure.dpi']=300
```

```python
from math import floor
s_const, s_var = 68269+17158, 6000
c_const, c_var = 0, 17271+1196
cli_max = 25
serv = np.array([s_const + s_var*i for i in range(cli_max+1)])
pc = np.array([c_const + c_var*i  for i in range(cli_max+1)])
eco = 100*(pc-serv)/serv
```

```python
fmt = "{} & {:.0f} & {:.0f} & {:.0f} \\\\ \n"
with open('../tex/eco.tex', 'w') as f:
    f.write(fmt.format('Толстые клиенты, руб', pc[15],pc[20],pc[25]))
    f.write(fmt.format('Cервер и тонкие клиенты, руб', serv[15],serv[20],serv[25]))
    f.write('\midrule\n')
    f.write(fmt.format('Экономия, \\%', eco[15],eco[20],eco[25]))
```

```python
plt.xlim(0,cli_max)
plt.ylim(0, 500)
plt.grid(True)

plt.plot(serv/1000, '-o')
plt.plot(pc/1000, '-o')

plt.xlabel('Клиенты', fontsize=10)
plt.ylabel('Стоимость, тыс. руб.', fontsize=10)
plt.savefig('../img/price_chart.png')
plt.clf()
```

```python
 a = np.array([
    [52.1  , 20.2 , 25.0 , 20.7],
    [51.0  , 30.5 , 30.2 , 29.4],
    [31.7  , 22.9 , 43.6 , 21.4],
    [7.6  ,  5.6 , 6.8  , 5.6],
    [105.0 , 44.6 , 55.7 , 43.2]
 ])
tests = ["Графика", "Процессор", "Ввод-вывод", "Рендеринг", "Симуляция"]
X = range(1,len(a[0])+1)
for index, param in enumerate(a):
    plt.bar(X, param, width = 0.5, bottom=sum(a[i] for i in range(index)))
plt.legend(labels=tests)
plt.xticks(X)
plt.xlabel('Сценарий', fontsize=10)
plt.ylabel('Время, с', fontsize=10)
plt.savefig('../img/perf_chart.png')
```

```python
fmt = "{} & {:.1f} & {:.1f} & {:.1f} & {:.1f} \\\\ \n"
with open('../tex/perf.tex', 'w') as f:
    for index, line in enumerate(a):
        f.write(fmt.format(tests[index], *line).replace('.', ','))
```
