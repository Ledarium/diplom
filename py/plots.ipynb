```python
import numpy as np
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
import matplotlib as mpl
mpl.rcParams['figure.dpi']=300
```

```python
from math import floor
s_const, s_var = 68269+17158, 6000
c_const, c_var = 0, 17271+1196
cli_max = 25
serv = np.array([s_const + s_var*i for i in range(cli_max+1)])
pc = np.array([c_const + c_var*i  for i in range(cli_max+1)])
eco = 100*(pc-serv)/serv
```

```python
fmt = "{} & {:,.0f} & {:,.0f} & {:,.0f} \\\\ \n"
with open('../tex/eco.tex', 'w') as f:
    f.write(fmt.format('Толстые клиенты', pc[15],pc[20],pc[25]).replace(',', ' '))
    f.write(fmt.format('Cервер и тонкие клиенты', serv[15],serv[20],serv[25]).replace(',', ' '))
    f.write('\midrule\n')
    f.write(fmt.format('Экономия, \\%', eco[15],eco[20],eco[25]))
```

```python
plt.xlim(0,cli_max)
plt.ylim(0, 500)
plt.grid(True)

plt.plot(serv/1000, '-o')
plt.plot(pc/1000, '-o')

plt.xlabel('Клиенты', fontsize=10)
plt.ylabel('Стоимость, тыс. руб.', fontsize=10)
plt.savefig('../img/price_chart.png')
```
